```python
def pairplot(df, target):
    ncol, nrow = 7, df.shape[1] // 7 + (df.shape[1] % 7 > 0)
    plt.figure(figsize=(ncol * 4, nrow * 4))

    for i, feature in enumerate(df.columns):
        plt.subplot(nrow, ncol, i + 1)
        plt.scatter(df[feature], target, s=10, marker='o', alpha=.6)
        plt.xlabel(feature)
        if i % ncol == 0:
            plt.ylabel('target')
```

Постройте график зависимости $R^2$ для обучающей и тестовой (на кросс-валидации) выборок в зависимости от значения гиперпараметра. Для Lasso также постройте график зависимости количества ненулевых весов.

```python
from sklearn.model_selection import cross_validate, KFold

train_indices = []
test_indices = []
kfolds = KFold(n_splits=5, random_state=1, shuffle=True)
cv_folds = kfolds.split(x_train)
for train_index, test_index in cv_folds:
    train_indices.append(train_index)
    test_indices.append(test_index)
    
    
mean_train_cvR = [0]*len(params['alpha'])
mean_test_cvR = [0]*len(params['alpha'])
mean_train_cvL = [0]*len(params['alpha'])
mean_test_cvL = [0]*len(params['alpha'])
for i in range(kfolds.get_n_splits(x_train)):

    modR = GridSearchCV(Ridge(), params, cv=[[train_indices[i], test_indices[i]]], scoring='r2',
                       return_train_score=True)
    modL = GridSearchCV(Lasso(), params, cv=[[train_indices[i], test_indices[i]]], scoring='r2',
                       return_train_score=True)
    modR.fit(x_train, y_train)
    modL.fit(x_train, y_train)

    mean_train_cvR = list(map(lambda x,y: x+y, mean_train_cvR, modR.cv_results_['mean_train_score']))
    mean_test_cvR = list(map(lambda x,y: x+y, mean_test_cvR, modR.cv_results_['mean_test_score']))
    mean_train_cvL = list(map(lambda x,y: x+y, mean_train_cvL, modL.cv_results_['mean_train_score']))
    mean_test_cvL = list(map(lambda x,y: x+y, mean_test_cvL, modL.cv_results_['mean_test_score']))

mean_train_cvR = list(map(lambda x: x/kfolds.get_n_splits(x_train), mean_train_cvR))
mean_test_cvR = list(map(lambda x: x/kfolds.get_n_splits(x_train), mean_test_cvR))
mean_train_cvL = list(map(lambda x: x/kfolds.get_n_splits(x_train), mean_train_cvL))
mean_test_cvL = list(map(lambda x: x/kfolds.get_n_splits(x_train), mean_test_cvL))


lasso_coefs = []
for i in params['alpha']:
    model = Lasso(alpha=i).fit(x_train, y_train)
    lasso_coefs.append(np.count_nonzero(model.coef_))

fig = plt.figure(figsize=(20,20))

ax1 = fig.add_subplot(321)
ax1.semilogx(params['alpha'], mean_test_cvR, label=u"Тестовая выборка")
ax1.semilogx(params['alpha'], mean_train_cvR, label=u"Обучающая выборка")
plt.title(u"Ridge()")
plt.xlabel(u"Alphas")
plt.ylabel(u"R^2 метрика")
plt.grid()
plt.legend()

ax2 = fig.add_subplot(322)
ax2.semilogx(params['alpha'], mean_test_cvL, label=u"Тестовая выборка")
ax2.semilogx(params['alpha'], mean_train_cvL, label=u"Обучающая выборка")
plt.title(u"Lasso()")
plt.xlabel(u"Alphas")
plt.ylabel(u"R^2 метрика")
plt.grid()
plt.legend()

ax3 = fig.add_subplot(323)
ax3.semilogx(params['alpha'], lasso_coefs)
plt.title(u'Зависимость кол-ва ненулевых весов Lasso()')
plt.xlabel(u'Alphas')
plt.ylabel(u'Кол-во ненулевых весов')
plt.grid()

plt.show()
```

В какой момент происходит недообучение? Почему?

__Ответ__: Недообучение начинает происходить при увеличение нашего альфа для регуляризаторов `L1`и `L2`. При увеличении альфы регуляризатора `L1`, происходит зануление многих признаков. Картину с недообучением можно также наблюдать и у `Ridge()` регресси при увеличении коэффициента регуляризации альфа

Для одной комбинаций метода масштабирования и линейной регрессии постройте зависимость метрики $R^2$ на кросс-валидации (`GridSearchCV.cv_results_['mean_test_score']`) от значения параметра регуляризации для различных степеней полиномов в `PolynomialFeatures`.

```python
r2_cross = []
for degree in range(1, 3):
    fin_pipe.set_params(poly__degree=degree)
    fin_pipe.fit(x_train, np.log(y_train))
    val = fin_pipe.named_steps['Ridge_r'].cv_results_['mean_test_score']
    r2_cross.append(val)

```
